```python
from sklearn import datasets
import numpy as np
from sklearn import preprocessing
iris=datasets.load_iris()
X=iris.data[:,[2,3]]
y=iris.target
std_scale=preprocessing.StandardScaler().fit(X)
X_std=std_scale.transform(X)
minmax_scale=preprocessing.MinMaxScaler().fit(X)
X_minmax=minmax_scale.transform(X)
```

```python

# Load the Iris dataset
iris = datasets.load_iris()

# Create a DataFrame from the Iris data and target
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                       columns=iris['feature_names'] + ['species'])

# Map numerical target values to corresponding species names
iris_df['species'] = np.where(iris_df['species'] == 0.0, 'setosa',
                              np.where(iris_df['species'] == 1.0, 'versicolor', 'virginica'))

# Remove any spaces from column names
iris_df.columns = iris_df.columns.str.replace(' ', '')

# Display summary statistics of the Iris DataFrame
iris_desc = iris_df.describe()
print(iris_desc)
```

```python
print(iris_df['species'].value_counts())
```

```python
iris_df.hist()
plt.suptitle("Histogram", fontsize=16)
plt.show()
```

```python
iris_df.boxplot()
plt.title("Bar Plot", fontsize=16)
plt.show()
```

```python
print(iris_df.groupby(by="species").mean())
iris_df.groupby(by="species").mean().plot(kind="bar")
plt.title('Class vs Measurements')
plt.ylabel('mean measurement(cm)')
plt.xticks(rotation=0)
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```

```python
numeric_columns = ['sepallength(cm)', 'sepalwidth(cm)', 'petallength(cm)', 'petalwidth(cm)']
numeric_df = iris_df[numeric_columns]

# Calculate the correlation matrix
corr = numeric_df.corr()
print(corr)
import statsmodels.api as sm
sm.graphics.plot_corr(corr,xnames=list(corr.columns))
```
